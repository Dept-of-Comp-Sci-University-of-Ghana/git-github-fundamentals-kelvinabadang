WEB SCRAPING

```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import requests
```

```python
base_url = 'https://www.gutenberg.org/files/15784/15784-0.txt'


```

```python
response = requests.get(base_url)
type (response)
response
```

```python
soup_dos = BeautifulSoup(response.content, 'html.parser')
len(soup_dos)
dos_text = soup_dos.get_text()
type(dos_text)
dos_text[:100]
```

```python
import re
a ='Who knew Johnny Depp was an undercover police officer (with Richard Greico)!'
ds = 'd/w+'
re.findall(ds, a)
ds = '[dD]\w+'
re.findall(ds,a)
words = re.findall('w+', dos_text)
```

DATA PROCESSING USING NLP

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(dos_text)
tokens[:8]
```

```python
words = []
for word in tokens:
    words.append(word.lower())
words[:10]    
   
```

```python
from nltk.corpus import stopwords
set(stopwords.words('english'))
stop_words = set(stopwords.words('english'))
filter_text = [word for word in words if not word is stop_words ]
filter_text[:10]
```

```python
import nltk
text = nltk.Text(filter_text)
text[:10]
fWords = nltk.FreqDist(text)
type(fWords)
fWords.most_common(10)
fWords['blood']
plt.figure(figsize = (9,7))
fWords.plot(30)

plt.figure()
fWords.plot(10, cumulative=True)
```
